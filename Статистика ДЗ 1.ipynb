```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
# распределение Пуассона

poisson_rv = sts.poisson(mu = 3)

x = np.linspace(0,10,10000)

cdf = poisson_rv.cdf(x)
pmf = poisson_rv.pmf(x)


plt.plot(x, pmf, 'o')
plt.plot(x, cdf)

plt.legend(['pmf - Probability mass function',  'cdf - Cumulative density function'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
import seaborn as sns
data_poisson = sts.poisson.rvs(mu=3, size=10000)
ax = sns.distplot(data_poisson,
                  bins=30,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 20,'alpha':0.6})
№ax.set(xlabel='Poisson Distribution', ylabel='Frequency')
```

```python
# Экспоненциальное распределение

expon_rv = sts.expon
x = np.linspace(0,4,100)

cdf = expon_rv.cdf(x)
pdf = expon_rv.pdf(x)

plt.plot(x, pdf, lw =3, alpha = 0.5)
plt.plot(x, cdf)


plt.legend(['pdf - Probability density function',  'cdf - Cumulative density function'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
# распределение Лапласа, которого не было в лекции

laplace_rv = sts.laplace
x = np.linspace(-2,2,100)

cdf = laplace_rv.cdf(x)
pdf = laplace_rv.pdf(x)

plt.plot(x, pdf)
plt.plot(x, cdf)

plt.legend(['pdf - Probability density function',  'cdf - Cumulative density function'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```
