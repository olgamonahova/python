```python
# 2.распределение Пуассона

poisson_rv1 = sts.poisson(mu = 1)
poisson_rv2 = sts.poisson(mu = 2)
poisson_rv3 = sts.poisson(mu = 3)

x = np.linspace(0,10,100)

pmf1 = poisson_rv1.pmf(x)
pmf2 = poisson_rv2.pmf(x)
pmf3 = poisson_rv3.pmf(x)


plt.plot(x, pmf1, 'o')
plt.plot(x, pmf2, 'o')
plt.plot(x, pmf3, 'o')

plt.legend(['mu = 1','mu = 2','mu = 3'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(0,10,100)

cdf1 = poisson_rv1.cdf(x)
cdf2 = poisson_rv2.cdf(x)
cdf3 = poisson_rv3.cdf(x)


plt.plot(x, cdf1)
plt.plot(x, cdf2)
plt.plot(x, cdf3)

plt.legend(['mu = 1','mu = 2','mu = 3'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
# Экспоненциальное распределение

expon_rv1 = sts.expon(scale = 2)
expon_rv2 = sts.expon(scale = 1)
expon_rv3 = sts.expon(scale = 0.5)


x = np.linspace(0,4,100)

#cdf = expon_rv.cdf(x)
pdf1 = expon_rv1.pdf(x)
pdf2 = expon_rv2.pdf(x)
pdf3 = expon_rv3.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.plot(x, pdf3)

#plt.plot(x, cdf)

plt.legend(['lamda 0.5, pdf',  'lamda 1, pdf' , 'lambda 2, pdf'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
cdf1 = expon_rv1.cdf(x)
cdf2 = expon_rv2.cdf(x)
cdf3 = expon_rv3.cdf(x)

plt.plot(x, cdf1)
plt.plot(x, cdf2)
plt.plot(x, cdf3)


plt.legend(['lamda 0.5, cdf',  'lamda 1, cdf' , 'lambda 2, cdf'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
# распределение Лапласа, которого не было в лекции

laplace_rv11 = sts.laplace(loc = 1, scale = 1)
laplace_rv12 = sts.laplace(loc = 1, scale = 2)
laplace_rv22 = sts.laplace(loc = 2, scale = 2)

x = np.linspace(-2,4,100)

pdf1 = laplace_rv11.pdf(x)
pdf2 = laplace_rv12.pdf(x)
pdf3 = laplace_rv22.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.plot(x, pdf3)


plt.legend(['loc = 1, scale = 1',  'loc = 1, scale = 2', 'loc = 2, scale = 2'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(-2,4,100)

cdf1 = laplace_rv11.cdf(x)
cdf2 = laplace_rv12.cdf(x)
cdf3 = laplace_rv22.cdf(x)

plt.plot(x, cdf1)
plt.plot(x, cdf2)
plt.plot(x, cdf3)


plt.legend(['loc = 1, scale = 1',  'loc = 1, scale = 2', 'loc = 2, scale = 2'])
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```
