```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
def get_formula(_cols):
    return 'diagnosis ~ ' + " + ".join(_cols)
```

```python
_iteration_cols = x_train.columns

columns = np.append(_iteration_cols,'diagnosis')

train_df = pd.DataFrame(np.c_[x_train_scaled, y_train], columns=columns)
get_formula(_iteration_cols)
```

# Backward selection

## Итерация 1

```python
lm = smf.ols(get_formula(_iteration_cols), train_df).fit()
rss = np.sum(lm.resid ** 2)
rmse = np.sqrt(rss / len(lm.resid))
print("RMSE:", rmse)
print("R^2:", lm.rsquared)

lm_res_0 = lm.summary()
lm_res_0.tables[1]
```

9. Решить задачу бинарной классификации и предсказать переменную ”Diagnosis ” протестировав как минимум 2 алгоритма. Использовать те алгоритмы, которые позволяют предсказать вероятность класса (proba). Рассчитать и вывести вероятность каждого класса.
10. Проверить качество классификации с использованием следующих метрик: Accuracy, F1-Score, Precision, Recall

```python
x_train, x_test, y_train, y_test = train_test_split(X, y,stratify=y, test_size=0.3, random_state=42)
```

```python
# по совокупности результатов наиболее точная модель для предсказания диагноза - xgboost, у него
#одинаково высокий score для train и test, минимальный RSS, максимальный R^2

xgb = xgboost.XGBClassifier(max_depth=4, n_jobs=-1)
xgb.fit(x_train, y_train)
y_train_predict = xgb.predict_proba(x_train)[:, 1]
y_test_predict = xgb.predict_proba(x_test)[:, 1]

# расчет привычных метрик
roc_auc_train = np.round(roc_auc_score(y_train, y_train_predict), 2)
roc_auc_test = np.round(roc_auc_score(y_test, y_test_predict), 2)

# расчет метрик из задания
accuracy = np.round(accuracy_score(y_test, y_test_predict.round()),2)
f1 = np.round(f1_score(y_test, y_test_predict.round()),2)
presicion = np.round(precision_score(y_test, y_test_predict.round()),2)
recall = np.round(recall_score(y_test, y_test_predict.round()),2)

print("Train score: ", roc_auc_train)
print("Test score: ", roc_auc_test)
print("Accuracy:" , accuracy)
print("F1-score:" , f1)
print("Precision_score:" , presicion)
print("Recall_score:" , recall)


rss = np.round(sum((y_test_predict-y_test) ** 2),2)
R2 = np.round(r2_score(y_test, y_test_predict),2)

print("RSS test:", rss)
print("R^2 test:", R2)

# на всякий случай посмотрим кросс-валидацию, чтобы исключить случайность полученного результата
scores_train = cross_validate(
    xgb, x_train, y_train, cv = 5, scoring=('roc_auc'), return_train_score=True
)

print('mean cross val train:',np.round(scores_train.get('test_score').mean(),2))

scores_test = cross_validate(
    xgb, x_test, y_test, cv = 5, scoring=('roc_auc'), return_train_score=True
)

print('mean cross val test:', np.round(scores_test.get('test_score').mean(),2))
```

```python
# Линейная регрессия оказалась хуже, чем xgboost, хотя в целом показала хороший результат

lr = LinearRegression()

lr.fit(x_train, y_train)
y_train_predict = lr.predict(x_train)
y_test_predict = lr.predict(x_test)


roc_auc_train = np.round(roc_auc_score(y_train, y_train_predict), 2)
roc_auc_test = np.round(roc_auc_score(y_test, y_test_predict), 2)

accuracy = np.round(accuracy_score(y_test, y_test_predict.round()),2)
f1 = np.round(f1_score(y_test, y_test_predict.round()),2)
presicion = np.round(precision_score(y_test, y_test_predict.round()),2)
recall = np.round(recall_score(y_test, y_test_predict.round()),2)

print("Train score: ", roc_auc_train)
print("Test score: ", roc_auc_test)
print("Accuracy:" , accuracy)
print("F1-score:" , f1)
print("Precision_score:" , presicion)
print("Recall_score:" , recall)

rss = np.round(sum((y_test_predict-y_test) ** 2),2)
R2 = np.round(r2_score(y_test, y_test_predict),2)

print("RSS test:", rss)
print("R^2 test:", R2)

scores_train = cross_validate(
    lr, x_train, y_train, cv = 5, scoring=('roc_auc'), return_train_score=True
)

print('mean cross val train:',np.round(scores_train.get('test_score').mean(),2))

scores_test = cross_validate(
    lr, x_test, y_test, cv = 5, scoring=('roc_auc'), return_train_score=True
)

print('mean cross val test:', np.round(scores_test.get('test_score').mean(),2))
```

```python
# у модели сильно проседают метрики F1-score: 0.92 и Recall_score: 0.86
# т.е. модель делает много ложноотрицательных предсказаний 

rfc = RandomForestClassifier(n_estimators=10, max_depth=10)

rfc.fit(x_train, y_train)
y_train_predict = rfc.predict_proba(x_train)[:, 1]
y_test_predict = rfc.predict_proba(x_test)[:, 1]

roc_auc_train = np.round(roc_auc_score(y_train, y_train_predict), 2)
roc_auc_test = np.round(roc_auc_score(y_test, y_test_predict), 2)

accuracy = np.round(accuracy_score(y_test, y_test_predict.round()),2)
f1 = np.round(f1_score(y_test, y_test_predict.round()),2)
presicion = np.round(precision_score(y_test, y_test_predict.round()),2)
recall = np.round(recall_score(y_test, y_test_predict.round()),2)

print("Train score: ", roc_auc_train)
print("Test score: ", roc_auc_test)
print("Accuracy:" , accuracy)
print("F1-score:" , f1)
print("Precision_score:" , presicion)
print("Recall_score:" , recall)

rss = np.round(sum((y_test_predict-y_test) ** 2),2)
R2 = np.round(r2_score(y_test, y_test_predict),2)

print("RSS test:", rss)
print("R^2 test:", R2)

scores_train = cross_validate(
    rfc, x_train, y_train, cv = 5, scoring=('roc_auc'), return_train_score=True
)

print('mean cross val train:',np.round(scores_train.get('test_score').mean(),2))

scores_test = cross_validate(
    rfc, x_test, y_test, cv = 5, scoring=('roc_auc'), return_train_score=True
)

print('mean cross val test:', np.round(scores_test.get('test_score').mean(),2))
```
