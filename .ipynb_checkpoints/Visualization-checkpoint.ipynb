## Часть 1. Matplotlib

```python
import pandas as pd

df = pd.read_csv('Players.csv')
print(df.shape)
```

```python
df.info()
```

```python
df = df.dropna()
print(df.shape)
```

```python
df.head()
```

```python
df['born'] = df.born.astype('int64')
```

```python
df.head()
```

```python
# посмотрим как менялся средний рост по годам
df1 = df[['height','born']].groupby('born').mean().reset_index()
df1.head()
```

```python
# распределение игроков по росту
ax = df.height.hist(bins = 10)
ax.set_title('Player Height Distribution')
ax.set_xlabel('Height')
ax.set_ylabel('Number of players')
```

```python
#динамика среднего роста по годам
df1.style.bar()
```

## Часть 2. Seaborn

```python
import seaborn as sns
```

```python
# распределение по росту игроков
%config InlineBackend.figure_format = 'svg' 
sns.distplot(df.height);
```

```python
# график для проверки гипотезы, есть ярко выраженный тренд изменения среднего роста по годам или нет, 
# ответ, по графикам видна только сильная зависимость роста и веса
# c svg pairplot браузер начинает тормозить
%config InlineBackend.figure_format = 'png' 
sns_plot = sns.pairplot(
    df[['born', 'height', 'weight']]);
sns_plot.savefig('pairplot.png')
```

```python
# как меняется рост игроков по годам
# ответ: средний рост игроков постепенно увеличивается 
sns.jointplot(x='born', y='height', 
              data=df, kind='reg');
```

```python
df.birth_state.value_counts().head(5).index.values
```

```python
# проверяем, есть ли какая-то зависимость между ростом игроков и штатом, где они родились
# разброс получился незначительным, география на рост сильно не влияет
top_state = df.birth_state.value_counts().sort_values(ascending = False).head(5).index.values
sns.boxplot(x="birth_state", y="height", 
            data=df[df.birth_state.isin(top_state)])
```

# Часть 3. Plotly

```python
from plotly.offline import init_notebook_mode, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
born_df = df.groupby('born')[['Player']].count()
born_df.head()
```

```python
def plotly_line_plot(df, title = ''):
    data = []
    
    for column in df.columns:
        trace = go.Scatter(
            x = df.index,
            y = df[column],
            mode = 'lines',
            name = column
        )
        data.append(trace)
    
    layout = dict(title = title)
    fig = dict(data = data, layout = layout)
    iplot(fig, show_link=False)
```

```python
#интерактивный график

plotly_line_plot(
    born_df
    ,
    title = 'Number of Players'
)
```
